```python
##############################################   Задание 1.1   #################################################################

######## Постройте график с помощью plt.scatter, где по горизонтальной оси будут значения Х, а по вертикальной оси - значения Y.
######## Добавьте название графика "Мой первый график"
######## Добавьте подпись горизонтальной оси "Ряд Х"
######## Добавьте подпись вертикальной оси "Ряд Y"
######## Добавьте сетку на график с помощью функции grid


from matplotlib import pyplot as plt

x = [1,2,3,4,5,6,7,8,9]
y = [8,5,6,10,18,26,18,22,23]
```

```python
plt.title('Мой первый график')
plt.xlabel('Ряд Х')
plt.ylabel('Ряд Y')

plt.grid()
plt.scatter(x,y)
plt.show()
```

```python
##############################################   Задание 1.2   #################################################################

######## Постройте график с помощью plt.plot, где по горизонтальной оси будут значения Х, а по вертикальной оси - значения Y.
######## Повторите все те же самые действия по оформлению графика, как в прошлом пункте.
######## Установите размер графика 7*7 с помощью параметра figsize функции figure
plt.figure(figsize=(7,7))

plt.title('Мой первый график')
plt.xlabel('Ряд Х')
plt.ylabel('Ряд Y')

plt.grid()
plt.plot(x,y)
```

```python
##############################################   Задание 1.3   #################################################################

##### Добавьте на визуализацию из прошлого пункта вертикальную линию красного цвета в точке 6 
##### Добавьте также пунктирную вертикальную линию черного цвета в точке 2
##### Сделайте кривую y(x) зеленого цвета
##### Добавьте легенду для максимума и минимума в правый нижний угол

plt.figure(figsize=(7,7))

plt.title('Мой первый график')
plt.xlabel('Ряд Х')
plt.ylabel('Ряд Y')

plt.grid()
plt.plot(x,y,c='g',)
plt.axvline(x=6, color='r', linestyle='-', linewidth=1, label='max Y')
plt.axvline(x=2, color='k', linestyle='--', linewidth=1, label='min X')

plt.legend(loc='lower right')
```

```python
##############################################   Задание 1.4   #################################################################

##### Добавьте на график еще одную кривую z = [6,9,13,13,16,12,20,18,17] и сделайте ее фиолетовго цвета
##### Теперь должно быть 4 вертикальные линии - max(Y), min(Y), max(Z), min(Z). 
##### Подпишите их на легенде и выберите цвета для них.Максимумы должны быть пунктирные, а минимумы цельные

z = [6,9,13,13,16,12,20,18,17]
plt.figure(figsize=(7,7))

plt.title('Мой первый график')
plt.xlabel('Ряд Х')
plt.ylabel('Ряд Y,Z')

plt.grid()
plt.plot(x,y,c='g')
plt.plot(x,z,c='m')
plt.axvline(x=6, color='r', linestyle='-', linewidth=1, label='max Y')
plt.axvline(x=2, color='k', linestyle='--', linewidth=1, label='min Y')
plt.axvline(x=7, color='y', linestyle='-', linewidth=1, label='max Z')
plt.axvline(x=1, color='b', linestyle='--', linewidth=1, label='min Z')

plt.legend(loc='lower right')
```

```python
##############################################   Задание 1.5   #################################################################

###### Воспроизведите тот же график, что и в прошлом пункте, используя ax и subplots

fig, ax = plt.subplots(figsize=(20,7))

ax.set_title('Мой график ')
ax.set_xlabel('Ряд Х')
ax.set_ylabel('Ряд Y,Z')

ax.grid()

ax.plot(x,y,c='g')

ax.plot(x,z,c='m')

ax.axvline(x=7, 
            color='y',
            linestyle='-', 
            linewidth=1,
            label = 'max z')

ax.axvline(x=1, 
            color='b',
            linestyle='-', 
            linewidth=1,
            label = 'min z')


ax.axvline(x=6, 
            color='r',
            linestyle='--', 
            linewidth=1,
            label = 'max y')


ax.axvline(x=2, 
            color='k',
            linestyle='-', 
            linewidth=1,
            label = 'min y')

ax.legend(loc = 4)
```

```python
##############################################   Задание 1.6   #################################################################

##### Добавьте на правую ось гистограмму, которая будет показывать разницу между Y и Z
##### Сделайте эту гистограмму полупрозрачной с помощью аргумента alpha
##### Добавьте подпись для гистограммы на отдельный элемент легенды.
##### Расположите основную легенду справа-снизу, а легенду вспомогательной оси справа-сверху

########  Hint: используйте twinx() для создания дополнительной оси
fig, ax = plt.subplots(figsize=(20,7))
ax2 = ax.twinx ()

ax.set_title('Мой график ')
ax.set_xlabel('Ряд Х')
ax.set_ylabel('Ряд Y,Z')

ax.grid()

ax.plot(x,y,c='g')

ax.plot(x,z,c='m')

ax2.bar([1,2,3,4,5,6,7,8,9],[2,-4,7,-3,2,14,-2,4,6], color ='c', label='"difference between y and z"', alpha =0.5)

ax.axvline(x=7, 
            color='y',
            linestyle='-', 
            linewidth=1,
            label = 'max z')

ax.axvline(x=1, 
            color='b',
            linestyle='-', 
            linewidth=1,
            label = 'min z')


ax.axvline(x=6, 
            color='r',
            linestyle='--', 
            linewidth=1,
            label = 'max y')


ax.axvline(x=2, 
            color='k',
            linestyle='-', 
            linewidth=1,
            label = 'min y')

ax.legend(loc = 4)
ax2.legend('difference')
```

### Задание 2. Поработаем с визуализацией датасета

```python
##############################################   Задание 2.1   #################################################################

##### Импортните csv-файл под названием Purchase_List
##### Приведите date_purchase к формату даты с помощью функции to_datetime
##### Создайте колонку датафрейма с месяцесем оплат (числом, от 1 до 12)
##### Сделайте группировку: количество оплат и сумма оплат для каждого месяца (создайте новый группировочный датафрейм)
##### Постройте визуализацию: для каждого месяца покупки по левой оси постройте сумму оплат, а по правой - количество оплат
##### Не забудьте оформление графика (заглавие, подпись осей, раздельная легенда, сетка)

import pandas as pd
import datetime as dt

df = pd.read_csv('Purchase_List.csv')
df['month_purchase'] = pd.to_datetime(df['date_purchase']).dt.month
df_group = df.groupby('month_purchase')['amt_pay'].agg(['count', 'sum'])
df_group
```

```python
fig, ax = plt.subplots(figsize=(20,7))
ax2 = ax.twinx ()
ax.set_title('Purchase')
ax.set_xlabel('Month')
ax.set_ylabel('Purchase')

ax.grid()

ax.plot(df_group['sum'],c='r')

ax2.plot(df_group['count'], color ='b') 

ax.legend('Sum')
ax2.legend('Count')
```

```python
##############################################   Задание 2.2   #################################################################

##### Постройте scatter plot для среднего чека для каждого месяца.
##### Для этого создайте новую колонку в сгруппированном датафрейме с отношением суммы платежей к их количеству
##### Сохраните форматирование из предыдущих пунктов
##### Задайте нижнюю границу оси Y как ноль. Верхнюю границу придумайте самостоятельно для лучшего отображения информации.

```

```python
df_group['avg_pay']=df_group['sum']/df_group['count']
df_group
```

```python
plt.plot(df_group['avg_pay'],
        color='brown')
plt.title("Средний чек")
plt.grid()
plt.xlabel('Month')
plt.ylabel('Avg_pay')
```
